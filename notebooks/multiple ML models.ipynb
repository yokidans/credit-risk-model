# Fraud Detection System with Basel II Compliance

```python
# Enhanced Fraud Detection System 
# Import required libraries
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                            f1_score, roc_auc_score, confusion_matrix, 
                            precision_recall_curve, roc_curve, average_precision_score,
                            classification_report, brier_score_loss)
from sklearn.calibration import calibration_curve, CalibratedClassifierCV
from sklearn.preprocessing import RobustScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import make_pipeline as make_imb_pipeline
import xgboost as xgb
import lightgbm as lgb
import catboost as cb
import matplotlib.pyplot as plt
import seaborn as sns
import json
import joblib
import os
from datetime import datetime
import shap
import numpy as np
import pandas as pd
from scipy import stats
import optuna
from optuna.samplers import TPESampler
from optuna.integration import OptunaSearchCV

# Constants
RANDOM_STATE = 42
SMOTE_SAMPLING_STRATEGY = 0.1  # 10% minority class after resampling
MIN_PD = 0.001  # Minimum probability of default for Basel II
LGD = 0.45  # Loss Given Default
EAD_FACTOR = 1.0  # Exposure at Default factor

class ExperimentTracker:
    def __init__(self, experiment_name):
        self.experiment_name = experiment_name
        self.base_path = f"experiments/{experiment_name}"
        self._create_folder_structure()
        self.results = {}
        self.best_models = {}
        self.basel_results = {}
        
    def _create_folder_structure(self):
        os.makedirs(f"{self.base_path}/configs", exist_ok=True)
        os.makedirs(f"{self.base_path}/results", exist_ok=True)
        os.makedirs(f"{self.base_path}/artifacts", exist_ok=True)
        os.makedirs(f"{self.base_path}/models", exist_ok=True)
        os.makedirs(f"{self.base_path}/reports", exist_ok=True)
        
    def save_model(self, model, model_name):
        path = f"{self.base_path}/models/{model_name}.pkl"
        joblib.dump(model, path)
        return path
        
    def save_config(self, config, model_name):
        path = f"{self.base_path}/configs/{model_name}_config.json"
        with open(path, 'w') as f:
            json.dump(config, f)
        return path
        
    def save_metrics(self, metrics, model_name):
        self.results[model_name] = metrics
        path = f"{self.base_path}/results/{model_name}_metrics.json"
        with open(path, 'w') as f:
            json.dump(metrics, f)
        return path
        
    def save_basel_metrics(self, metrics):
        # Convert numpy types to native Python types for JSON serialization
        def convert(o):
            if isinstance(o, np.generic):
                return o.item()
            raise TypeError
        
        self.basel_results = metrics
        path = f"{self.base_path}/results/basel_metrics.json"
        with open(path, 'w') as f:
            json.dump(metrics, f, default=convert)
        return path
        
    def save_visualization(self, fig, name):
        path = f"{self.base_path}/artifacts/{name}.png"
        fig.savefig(path)
        plt.close(fig)
        return path
        
    def generate_report(self, dataset_stats):
        report = {
            "experiment_name": self.experiment_name,
            "date": datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
            "dataset_stats": dataset_stats,
            "model_performance": self.results,
            "basel_metrics": self.basel_results,
            "best_models": list(self.best_models.keys())
        }
        
        path = f"{self.base_path}/reports/final_report.json"
        with open(path, 'w') as f:
            json.dump(report, f, indent=4)
        return path

def feature_engineering(df):
    # Temporal features
    if 'TransactionStartTime' in df.columns:
        df['TransactionStartTime'] = pd.to_datetime(df['TransactionStartTime'])
        df['hour'] = df['TransactionStartTime'].dt.hour
        df['day_of_week'] = df['TransactionStartTime'].dt.dayofweek
        df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)
    
    # Transaction amount features
    if 'Amount' in df.columns:
        df['amount_log'] = np.log1p(df['Amount'])
        df['amount_zscore'] = stats.zscore(df['Amount'])
        df['amount_binned'] = pd.qcut(df['Amount'], q=5, labels=False)
    
    # Interaction features
    if all(col in df.columns for col in ['Amount', 'hour']):
        df['amount_hour_interaction'] = df['Amount'] * df['hour']
    
    return df

def prepare_data(df, target_col='FraudResult', test_size=0.2):
    df = feature_engineering(df)
    X = df.drop([target_col, 'TransactionId', 'TransactionStartTime'], axis=1, errors='ignore')
    y = df[target_col]
    
    categorical_cols = X.select_dtypes(include=['object', 'category']).columns
    numerical_cols = X.select_dtypes(include=['number']).columns
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', RobustScaler(), numerical_cols),
            ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
        ])
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=RANDOM_STATE, stratify=y)
    
    sm = SMOTE(random_state=RANDOM_STATE, sampling_strategy=SMOTE_SAMPLING_STRATEGY)
    X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
    
    return X_train_res, X_test, y_train_res, y_test, preprocessor

def define_models(preprocessor):
    from optuna.distributions import IntDistribution, FloatDistribution, CategoricalDistribution
    
    base_models = {
        "RandomForest": {
            "model": RandomForestClassifier(random_state=RANDOM_STATE, class_weight='balanced'),
            "params": {
                'classifier__n_estimators': IntDistribution(100, 300, step=50),
                'classifier__max_depth': IntDistribution(5, 30, step=5),
                'classifier__min_samples_split': IntDistribution(2, 10, step=2),
                'classifier__max_features': CategoricalDistribution(['sqrt', 'log2'])
            }
        },
        "XGBoost": {
            "model": xgb.XGBClassifier(random_state=RANDOM_STATE, eval_metric='aucpr', 
                                     use_label_encoder=False),
            "params": {
                'classifier__n_estimators': IntDistribution(100, 300, step=50),
                'classifier__max_depth': IntDistribution(3, 9, step=2),
                'classifier__learning_rate': FloatDistribution(0.01, 0.1, log=True),
                'classifier__subsample': FloatDistribution(0.8, 1.0, step=0.1),
                'classifier__colsample_bytree': FloatDistribution(0.8, 1.0, step=0.1),
                'classifier__gamma': FloatDistribution(0, 0.2, step=0.1)
            }
        },
        "LightGBM": {
            "model": lgb.LGBMClassifier(random_state=RANDOM_STATE, objective='binary'),
            "params": {
                'classifier__n_estimators': IntDistribution(100, 300, step=50),
                'classifier__max_depth': IntDistribution(5, 30, step=5),
                'classifier__learning_rate': FloatDistribution(0.01, 0.1, log=True),
                'classifier__num_leaves': IntDistribution(31, 100, step=10),
                'classifier__feature_fraction': FloatDistribution(0.8, 1.0, step=0.1)
            }
        },
        "CatBoost": {
            "model": cb.CatBoostClassifier(
                random_state=RANDOM_STATE,
                verbose=0,
                allow_writing_files=False  # Prevents directory creation issues
            ),
            "params": {
                'classifier__iterations': IntDistribution(100, 500),
                'classifier__depth': IntDistribution(4, 10),
                'classifier__learning_rate': FloatDistribution(0.01, 0.3, log=True)
            }
        }
    }
    
    models = {}
    for name, config in base_models.items():
        models[name] = {
            "model": Pipeline([
                ('preprocessor', preprocessor),
                ('feature_selector', SelectFromModel(config["model"])),
                ('classifier', config["model"])
            ]),
            "params": config["params"]
        }
    
    return models

def generate_visualizations(model, X_test, y_test, model_name, tracker):
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
    ax.set_title(f'{model_name} Confusion Matrix')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')
    tracker.save_visualization(fig, f'{model_name}_cm')
    
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('ROC Curve')
    ax.legend()
    tracker.save_visualization(fig, f'{model_name}_roc')
    
    # Precision-Recall Curve
    precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
    pr_auc = average_precision_score(y_test, y_pred_proba)
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.plot(recall, precision, label=f'{model_name} (AP = {pr_auc:.2f})')
    ax.set_xlabel('Recall')
    ax.set_ylabel('Precision')
    ax.set_title('Precision-Recall Curve')
    ax.legend()
    tracker.save_visualization(fig, f'{model_name}_pr')
    
    # Calibration Curve
    fraction_of_positives, mean_predicted_value = calibration_curve(y_test, y_pred_proba, n_bins=10)
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.plot(mean_predicted_value, fraction_of_positives, "s-", label=model_name)
    ax.plot([0, 1], [0, 1], "k:", label="Perfectly calibrated")
    ax.set_xlabel("Mean predicted value")
    ax.set_ylabel("Fraction of positives")
    ax.set_title(f'{model_name} Calibration Plot')
    ax.legend()
    tracker.save_visualization(fig, f'{model_name}_calibration')
    
    # SHAP Feature Importance
    try:
        if hasattr(model, 'named_steps'):
            processed_X = model.named_steps['preprocessor'].transform(X_test)
            if hasattr(model.named_steps['preprocessor'], 'get_feature_names_out'):
                feature_names = model.named_steps['preprocessor'].get_feature_names_out()
            else:
                feature_names = [f'feature_{i}' for i in range(processed_X.shape[1])]
            
            explainer = shap.Explainer(model.named_steps['classifier'], processed_X[:100])
            shap_values = explainer(processed_X[:100])
        else:
            feature_names = X_test.columns
            explainer = shap.Explainer(model, X_test[:100])
            shap_values = explainer(X_test[:100])
        
        fig, ax = plt.subplots(figsize=(10, 8))
        shap.summary_plot(shap_values, feature_names=feature_names, plot_type="bar", show=False)
        ax.set_title(f'{model_name} Feature Importance')
        tracker.save_visualization(fig, f'{model_name}_shap')
        
        fig, ax = plt.subplots(figsize=(10, 8))
        shap.summary_plot(shap_values, feature_names=feature_names, show=False)
        ax.set_title(f'{model_name} SHAP Values')
        tracker.save_visualization(fig, f'{model_name}_shap_detailed')
    except Exception as e:
        print(f"Could not generate SHAP plots for {model_name}: {str(e)}")

def calculate_basel_metrics(y_true, y_pred_proba, amount, min_pd=MIN_PD, lgd=LGD, ead_factor=EAD_FACTOR):
    pd_values = np.maximum(y_pred_proba, min_pd)
    expected_loss = pd_values * lgd * (amount * ead_factor)
    actual_loss = amount[y_true == 1] * lgd
    
    risk_coverage = (float(np.sum(expected_loss[y_true == 1])) / float(np.sum(actual_loss))) if np.sum(actual_loss) > 0 else float('nan')
    total_expected_loss = float(np.sum(expected_loss))
    total_actual_loss = float(np.sum(actual_loss))
    capital_adequacy = (float(total_expected_loss / total_actual_loss)) if total_actual_loss > 0 else float('nan')
    rwa = float(np.sum(expected_loss) * 12.5)  # Basel II factor
    
    return {
        "risk_coverage": risk_coverage,
        "capital_adequacy": capital_adequacy,
        "total_expected_loss": total_expected_loss,
        "total_actual_loss": total_actual_loss,
        "risk_weighted_assets": rwa,
        "num_fraud_cases": int(np.sum(y_true)),
        "avg_expected_loss": float(np.mean(expected_loss)),
        "avg_actual_loss": float(np.mean(actual_loss)) if len(actual_loss) > 0 else 0.0
    }

def train_models(X_train, y_train, X_test, y_test, models, tracker):
    for model_name, model_config in models.items():
        print(f"\nTraining {model_name}...")
        
        study = optuna.create_study(direction='maximize', sampler=TPESampler(seed=RANDOM_STATE))
        optuna_search = OptunaSearchCV(
            model_config["model"],
            model_config["params"],
            cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE),
            n_trials=20,
            scoring='roc_auc',
            n_jobs=-1,
            verbose=1,
            study=study
        )
        
        optuna_search.fit(X_train, y_train)
        best_model = optuna_search.best_estimator_
        tracker.best_models[model_name] = best_model
        tracker.save_model(best_model, model_name)
        tracker.save_config(optuna_search.best_params_, model_name)
        
        y_pred = best_model.predict(X_test)
        y_pred_proba = best_model.predict_proba(X_test)[:, 1]
        
        metrics = {
            "accuracy": accuracy_score(y_test, y_pred),
            "precision": precision_score(y_test, y_pred),
            "recall": recall_score(y_test, y_pred),
            "f1": f1_score(y_test, y_pred),
            "roc_auc": roc_auc_score(y_test, y_pred_proba),
            "pr_auc": average_precision_score(y_test, y_pred_proba),
            "brier_score": brier_score_loss(y_test, y_pred_proba),
            "best_params": optuna_search.best_params_
        }
        
        tracker.save_metrics(metrics, model_name)
        generate_visualizations(best_model, X_test, y_test, model_name, tracker)

def run_experiment(df):
    experiment_name = f"experiment_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
    tracker = ExperimentTracker(experiment_name)
    
    X_train, X_test, y_train, y_test, preprocessor = prepare_data(df)
    
    dataset_stats = {
        "total_samples": len(df),
        "fraud_rate": df['FraudResult'].mean(),
        "features_used": X_train.shape[1],
        "train_samples": len(X_train),
        "test_samples": len(X_test),
        "resampled_train_samples": len(y_train)
    }
    
    models = define_models(preprocessor)
    train_models(X_train, y_train, X_test, y_test, models, tracker)
    
    basel_results = {}
    for model_name, model in tracker.best_models.items():
        y_pred_proba = model.predict_proba(X_test)[:, 1]
        basel_results[model_name] = calculate_basel_metrics(
            y_test, 
            y_pred_proba, 
            X_test['Amount'].values if 'Amount' in X_test.columns else np.ones(len(X_test))
        )
    
    tracker.save_basel_metrics(basel_results)
    tracker.generate_report(dataset_stats)
    
    print("\nModel Performance Summary:")
    for model_name, metrics in tracker.results.items():
        print(f"\n{model_name}:")
        for metric, value in metrics.items():
            if metric != "best_params":
                print(f"{metric}: {value:.4f}")
    
    print("\nBasel II Compliance Metrics:")
    for model_name, metrics in basel_results.items():
        print(f"\n{model_name}:")
        for metric, value in metrics.items():
            print(f"{metric}: {value:.4f}")
    
    return tracker

# Execute the experiment
if __name__ == "__main__":
    # Option 1: Load sample data (for demonstration)
    print("Loading sample data for demonstration...")
    from sklearn.datasets import make_classification
    import pandas as pd
    
    # Create synthetic data that resembles fraud detection data
    X, y = make_classification(
        n_samples=10000,
        n_features=20,
        n_informative=8,
        n_redundant=2,
        n_clusters_per_class=2,
        weights=[0.99, 0.01],
        random_state=RANDOM_STATE,
        flip_y=0.001
    )
    
    # Create a DataFrame with realistic column names
    df = pd.DataFrame(X, columns=[f'feature_{i}' for i in range(X.shape[1])])
    df['TransactionId'] = range(10000)
    df['TransactionStartTime'] = pd.date_range(start='2023-01-01', periods=10000, freq='min')
    df['Amount'] = np.exp(np.random.normal(3, 1, 10000)).round(2)  # Log-normal distributed amounts
    df['FraudResult'] = y
    
    print(f"Created sample dataset with {len(df)} records ({df['FraudResult'].mean():.2%} fraud rate)")
    
    # Option 2: Uncomment and modify to load your actual data
    # import pandas as pd
    # df = pd.read_csv('your_fraud_data.csv')
    
    # Run the experiment
    tracker = run_experiment(df)
```
